```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.plot(figsize=(25,5))
```

```python
df.plot(kind='area', alpha=0.5) # alpha is responsible for changing contrast in colors
```

```python
df.plot(kind ='scatter',x='sepal_length', y='sepal_width')
```

```python
#using 3 parameters
df.plot(kind ='scatter',x='sepal_length', y='sepal_width', c='petal_length', s=80)

#'x','y' is use to plot a point and 'c' is use to decide the color of the plot ; s is yhe size of dot
```

```python
df.plot(kind ='hexbin',x='sepal_length', y='sepal_width', gridsize = 10)
```

```python
sns.countplot(df['species'])
```

```python
df.head()
```

```python
sns.heatmap(df.corr(),annot=True)


plt.title("corr between variables",color='Blue',size=15)
plt.show()
```

```python
from mpl_toolkits import mplot3d
```

```python
x = np.linspace (-1,6, 30) #b/w 1 to 6 give 30 data points
y = np.linspace(-1,6, 30)
z = x+y
x,y,z
```

```python
ax = plt.axes(projection='3d') #plt alias matplotlib; axes shows white color grid entire axes, structure of graph
ax.plot3D(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
iris_df = plt.axes(projection='3d')
iris_df.plot3D(df['sepal_length'],df['sepal_width'],df['petal_length'])
iris_df.set_xlabel('sepal_length')
iris_df.set_ylabel('sepal_width')
iris_df.set_zlabel('petal_length')
```

```python
#bar graph
df.plot.bar(figsize = (20,10))

#or df.plot(kind='bar')
```

```python
#select particular sample data
df.iloc[10:21].plot.bar(figsize = (20,10))
```

```python
#horizontal bar graph
df.iloc[10:21].plot.barh(figsize = (20,10))
```

```python
# Histogram
df.plot.hist(figsize = (20,10))
```

```python
#histogram for single col
df['petal_length'].plot.hist(figsize = (20,10))
```

```python
#to construct histograms for all columns

df.hist(figsize = (20,10), color='g', alpha=0.9) #inplace of 'g' we can give color code too
```

```python
sns.relplot(x='sepal_length', y='sepal_width', col='species', data=df) 
```

```python
sns.catplot(x='sepal_width', y='sepal_length', data=df)
```

```python
sns.pairplot(df)
```
