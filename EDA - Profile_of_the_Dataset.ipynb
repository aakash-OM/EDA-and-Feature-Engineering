```python
#return only numerical columns in list format
num_col = [feature for feature in df.columns if df[feature].dtypes != 'object']
num_col
# we can place '0' also in place of object
```

```python
#return only categorical columns in list format
cat_col = [feature for feature in df.columns if df[feature].dtypes == 'object']
cat_col
```

```python
#filter out 1 entity that is max with number_of_employees

df[df['no_of_employees'] == max(df['no_of_employees'])]
```

```python
# fetch only that continent for which number_of_employees greator than 600000
df[df['no_of_employees'] > 600000]['continent']
```

```python
# Using + operator to combine two columns: region_of_employment + continent = emp_region
df["emp_region"] = df['region_of_employment'].astype(str) +", "+ df["continent"]
df
```

```python
df.loc[2:3]
#return 2nd and 3rd row
```

```python
df.iloc[2:3]
#only returns 2nd row
```

Both loc and iloc are row-first, column-second. This is the opposite of what we do in native Python, which is column-first, 
row-second.

```python
df.iloc[1:4, :4]
```

```python
df.iloc[[0, 1, 2], 2]
#by using list
```

```python
df.loc[2:4, 'emp_region']
```
