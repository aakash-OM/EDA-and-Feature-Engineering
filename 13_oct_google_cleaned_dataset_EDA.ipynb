# Distribution of Numerical Data

```python
for features in numeric:
    sns.kdeplot(df_num[features])
    plt.xlabel(features)
    plt.ylabel("count")
    plt.title("Numerical features plot \n \n", weight = 'bold')
    plt.show()
```

```python
#OR

plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric)):
    plt.subplot(4, 2, i+1)
    sns.kdeplot(x=df[numeric[i]],shade=True, color='g')
    plt.xlabel(numeric[i])
    plt.tight_layout()
```

Observation:-
  - 'Month', 'day' are Normally Distributed.
  - 'Rating', 'Year' are Left Skewed.
  - 'Reviews', 'Size', 'Installs', 'Price' are Right Skewed.

# Distribution of Categorical Features

```python
plt.figure(figsize=(35, 25))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = ['Type', 'Content Rating','Category']

for i in range(0, len(category)):
    plt.subplot(4, 2, i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

**Observation -**
- Family is most Popular category
- Free Applications are more as compare to paid.
- Content rating is more for everyone except particular entity
