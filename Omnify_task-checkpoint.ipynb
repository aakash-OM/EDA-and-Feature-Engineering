# **1. Most Profitable channel**

```python
df['Payment ($)'] = df['Payment ($)'].astype(str).str.replace('$', '')
```

```python
df.head()
```

**Step 1: converting Payments into numerical datatype**

```python
df['Payment ($)'] = pd.to_numeric(df['Payment ($)'], errors='coerce')
```

```python
#checking Data types
df.info()
```

```python
#checking NAN value
df.isnull().sum()
```

**Step 2: To calculate Most Profitable channel replace null value in 'Search Keyword' with Mode.**


```python
df['Search Keyword'] = df['Search Keyword'].fillna(df['Search Keyword'].mode()[0])
```

```python
df.isnull().sum()
```

**Step 3: Calculating Profit**

```python
df['Profit'] = df['Payment ($)'].fillna(0) - df['Cost ($)']
df['Profit']
```

```python
df['Profit'].isnull().sum()
```

**Step 4: Aggregate Profit by Channel**

```python
channel_profit = df.groupby('Campaign')['Profit'].sum()
```

**Step5: Find the Most Profitable Channel**

```python
most_profitable_channel = channel_profit.idxmax()
```

```python
print("The most profitable channel is:", most_profitable_channel)
```

**The most profitable channel is: EK_Generic_Swimming_UK**

**Representation by bar Graph**

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.bar(channel_profit.index, channel_profit.values)
plt.xlabel("Marketing Channel")
plt.ylabel("Total Profit")
plt.title("Profitability of Marketing Channels")
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels for better visibility
plt.axhline(0, color='black', linewidth=0.5)  # Add a horizontal line at y=0 for reference
plt.tight_layout()
plt.show()
```

# 3. Look for insights by analyzing data points on time series graphs

**1.Plotting Cost over time to analyze advertising expenses**

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Week'], df['Cost ($)'])
plt.xlabel('Week')
plt.ylabel('Cost ($)')
plt.title('Cost Over Time')
plt.tight_layout()
plt.show()
```

**2.Plotting Leads and Prospects over time to analyze their trends**

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Week'], df['Leads'], label='Leads')
plt.plot(df['Week'], df['Prospects'], label='Prospects')
plt.xlabel('Week')
plt.ylabel('Count')
plt.title('Leads and Prospects Over Time')
plt.legend()
plt.tight_layout()
plt.show()
```

**3.Plotting Leads and Impressions over time to analyze their trends**

```python
plt.figure(figsize=(15, 6))
plt.plot(df['Week'], df['Leads'], label='Leads')
plt.plot(df['Week'], df['Impressions'], label='Impressions')
plt.xlabel('Week')
plt.ylabel('Count')
plt.title('Leads and Impressions Over Time')
plt.legend()
plt.tight_layout()
plt.show()
```
